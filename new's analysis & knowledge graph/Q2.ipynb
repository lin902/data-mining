```python
def merge_dictionaries(file1, file2, merged_file):
    with open(file1, 'r', encoding='utf-8') as f1, open(file2, 'r', encoding='utf-8') as f2, \
            open(merged_file, 'w', encoding='utf-8') as merged_f:
        # 复制第一个词典的内容到合并文件
        for line in f1:
            merged_f.write(line)

        # 添加空行分隔两个词典
        merged_f.write('\n')

        # 复制第二个词典的内容到合并文件
        for line in f2:
            merged_f.write(line)

    print("Dictionaries merged successfully.")

```

```python
# 两个正向情感词典文件路径
file1 = '/content/drive/MyDrive/Colab_Notebooks/5002/dsaa5002_project/Q2/dict/formal_pos.txt'
file2 = '/content/drive/MyDrive/Colab_Notebooks/5002/dsaa5002_project/Q2/dict/unformal_pos.txt'

# 合并后的正向文件路径
merged_file = '/content/drive/MyDrive/Colab_Notebooks/5002/dsaa5002_project/Q2/dict/merged_pos.txt'

# 合并词典文件
merge_dictionaries(file1, file2, merged_file)
```

Step 2: Classifier

With 40,000 training observations and each word in the vectorizer's vocabulary counting acting as a predictor training could be slow.
This issue is exacerbated when using cross validation as we need fit the model multiple times!
We'll set our classifier CV parameters so as to help keep the training time down to around 30 seconds or so.



*   l2 penalty (e.g., Ridge)
*   10 iterations per fit (remember, logistic regression has no closed form solution for the betas!)


*   5-fold CV
*   random state of 0 (the fitting can be stochastic)



```python
# Instantiate our Classifier
clf = LogisticRegressionCV(penalty='l2',        # L2 penalty (Ridge)
    max_iter=10,         # 10 iterations per fit
    cv=5,                # 5-fold cross-validation
    random_state=0       # Random state for reproducibility
                           )
```

Step 3: Pipeline

```python
from sklearn.pipeline import make_pipeline, Pipeline

# Construct the pipeline
pipe = make_pipeline(chinese_tokenizer_transformer, vec, clf)

pipe.verbose = True
```

```python
import re
import numpy as np
from sklearn.preprocessing import FunctionTransformer
from nltk.corpus import stopwords
import nltk
import string
import thulac
# 全局 THULAC 对象
thu = thulac.thulac()  # 初始化 THULAC 分词器
def thulac_tokenizer(text):
    thu = thulac.thulac()  # 初始化 THULAC 分词器
    result = thu.cut(text, text=True)  # 对文本进行分词
    tokens = [word.split('_')[0] for word in result.split()]  # 去除词性信息
    return tokens  # 返回分词结果，以列表形式



# 中文标点符号
chinese_punctuation = '。，？！&#8203;``【oaicite:0】``&#8203;（）《》“”‘’.：；“”、{}￥%……&*-—()  '
alpha = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZＨ'
nltk.download('stopwords')

# 加载中文停用词表
stopwords_set = set(stopwords.words('chinese'))


def prepocess(text):
    tokens = jieba.lcut(str(text))
    # 移除停用词数字
    tokens_filtered = [word for word in text if word not in stopwords_set and not re.match(r'\d+', word) and word not in alpha and word not in chinese_punctuation]

    # 返回处理后的文本

    return ''.join(tokens_filtered)

# 创建 FunctionTransformer
chinese_tokenizer_transformer = FunctionTransformer(np.vectorize(prepocess), validate=False)#get list


# 使用 TfidfVectorizer 并指定 THULAC 分词器
vec = TfidfVectorizer(tokenizer=thulac_tokenizer, min_df=0.0001, max_df=0.9)

# Construct the pipeline
pipe = make_pipeline(chinese_tokenizer_transformer, vec, clf)

pipe.verbose = True
```
